# Exploratory Data Analysis

Now that we have understood about the problem and we have obtained `Framingham Heart Study` dataset, lets start the Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import plotly.tools as tls
import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff

py.init_notebook_mode(connected=True)
```

```python
chd_data = pd.read_csv('Data/data.csv')
print(chd_data.head())
```

#### Data Description:

```python
chd_data.describe()
```

## 1. Missing Values

Now let's see if the data have any missing feature values

```python
missing_data = chd_data.isna().sum()[chd_data.isna().sum()>0]

py.iplot([go.Bar(x=missing_data.index, y=missing_data)])
```

## 2. Data Distribution

Now let's look into the distribution of the features and distribution over the prediction target.

#### 2.A. Age Distribution

```python
age_fig = tls.make_subplots(subplot_titles=('Agewise Distribution', 'Target-wise Age Distribution'), rows=1, cols=2)

# Agewise Distribution
age_dist = ff.create_distplot([chd_data.age], ['Age'])

# Target-wise Age Distribution
age_chd_no = chd_data.age[chd_data.TenYearCHD==0]
age_chd_yes = chd_data.age[chd_data.TenYearCHD==1]
age_chd_dist = ff.create_distplot([age_chd_no, age_chd_yes], ['Normal', 'CHD Suffered'], bin_size=1)

age_fig.append_trace(age_dist.data[0], 1, 1)
age_fig.append_trace(age_dist.data[1], 1, 1)
age_fig.append_trace(age_chd_dist.data[0], 1, 2)
age_fig.append_trace(age_chd_dist.data[1], 1, 2)
age_fig.append_trace(age_chd_dist.data[2], 1, 2)
age_fig.append_trace(age_chd_dist.data[3], 1, 2)

py.iplot(age_fig)
```

#### 2.A. Prediction Target Distribution

```python
target_dist = chd_data.TenYearCHD.value_counts()

py.iplot([go.Bar(x=['No CHD', 'CHD  Occured'], y=target_dist)])
```
